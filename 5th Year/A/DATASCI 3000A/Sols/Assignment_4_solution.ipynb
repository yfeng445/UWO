### Q 1.12 - <span style="color:red">[15]</span> - Calculate confidence interval (based on 99% confidence level) for the mean of predictions by bootstrapping. For this purpose, code a bootstrap function that in each iteration, draws (according to the requirements of bootstrapping) a sample from the training set, fits a linear regression model to it, uses the model to make predictions over the test set, and stores the mean of the predictions in an array. Your function must take as input arguments: model, Xtrain, ytrain, Xtest, and numboot=100. The function must return only one object that is an array of the recorded averages in $\mathrm{euros}$ - and not $log(\mathrm{euros})$. The unit of the confidence interval must be $\mathrm{euros}$ too.

```python
def BootstrapPred(MODEL, X_train, y_train, X_test, numboot=100):
    y_pred = np.zeros(numboot)
    for i in range(numboot):
        X_fit = X_train.sample(Xtrain.shape[0], replace=True)
        y_fit = y_train[X_fit.index]
        y_pred[i]=np.exp(MODEL.fit(X_fit, y_fit).predict(X_test)).mean() # record bootstrap statistic
    return y_pred
```

```python
yhat_s         = np.exp(model.fit(Xtrain, ytrain).predict(Xtest))

mean_yhat_s    = yhat_s.mean()                               # sample statistic

mean_yhat_boot = BootstrapPred(model, Xtrain, ytrain, Xtest) # bootstrap statistic
```

```python
# This cell was not part of the asking - just an FYI:
bm = pd.DataFrame(data=BootstrapPred(model, Xtrain, ytrain, Xtest, 400)-mean_yhat_s, columns=['Values'])
ax = bm.Values.plot(kind='hist', bins=20)
bm.Values.plot(kind='kde', ax=ax, secondary_y=True)
plt.show()
```

```python
# Bootstrap with CL of 99%
boot_CL = 99/100

p_1 = (1-boot_CL)/2
p_2 = 1-p_1

boot_quant = np.quantile(mean_yhat_boot-mean_yhat_s, [p_1, p_2])

boot_ci = [((mean_yhat_s - boot_quant[1])*1e-6).round(3),
           ((mean_yhat_s - boot_quant[0])*1e-6).round(3)]
print("Boot confidence interval is",boot_ci, "- in million euros.")
```

### Q 1.13 - <span style="color:red">[6]</span> - Construct a 99% confidence interval using the Central Limit Theorem (again in $\mathrm{euros}$).

```python
# Calculate CI using CLT:

stderr = yhat_s.std() / np.sqrt(yhat_s.shape[0])

cl = 0.99
lower = mean_yhat_s - ss.norm.ppf(cl) * stderr
upper = mean_yhat_s + ss.norm.ppf(cl) * stderr

print('The confidence interval is [%.3f, %.3f] - in million euros.' % (lower*1e-6, upper*1e-6))
```

### Q 1.14 - <span style="color:red">[10]</span> - We want to see the effect of sample size ($n$) on the CI calculated from CLT. Write a `for` loop which in each iteration randomly samples from your "sample statistic" and calculates and stores the width of the corresponding CI in an array. Obviously, you should start from a small $n$ and increase it per iteration. After the loop, plot sample size (*i.e.*, $n$) against CI width and report your observation in one sentence.

```python
result = np.zeros(1)
for i in range(10, yhat_s.shape[0], 10):
    yhat_s_new = np.random.choice(yhat_s, size=i)
    stderr = yhat_s_new.std() / np.sqrt(yhat_s_new.shape[0])
    L = yhat_s_new.mean() - ss.norm.ppf(cl) * stderr
    U = yhat_s_new.mean() + ss.norm.ppf(cl) * stderr
    result = np.append(result, np.absolute(L-U))

plt.scatter(list(range(10, yhat_s.shape[0], 10)), result[1:])
plt.xlabel("Sample Size")
plt.ylabel("CI Width")
plt.show()
```

#### *Answer*

The more the number of data points the narrower the CI would get.
