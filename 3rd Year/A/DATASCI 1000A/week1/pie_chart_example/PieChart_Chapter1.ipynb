```python
import pandas as pd

data = pd.read_csv (r'/Users/camiladesouza/OneDrive - The University of Western Ontario/DS1000/textbook_datasets_csv/chapter_01/eg01-02majors.csv') 
print(data)
```

```python
df = pd.DataFrame(data, columns = ['Field of Study', 'Percent']) 
```

```python
import matplotlib.pyplot as  plt
fig = df.plot.pie(y='Percent', figsize=(5,5),
                  labels=df['Field of Study'],legend=False,title="Percent of Students",ylabel='')
plt.show(fig)
```
